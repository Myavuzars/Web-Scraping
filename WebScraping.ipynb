```python
import requests
from bs4 import BeautifulSoup
```

```python
html_text= requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=').text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx') 
```

```python
#Printing the collected informations
for job in jobs:
  job_type= job.header.h2.a.text
  comp_name= job.find('h3', class_ = 'joblist-comp-name').text
  skills = job.find('span', class_='srp-skills').text.replace(' ','')
  job_pub_date = job.find('span', class_= 'sim-posted').text
  More_info = job.header.h2.a['href']
  year = job.find('ul', class_='top-jd-dtl clearfix').text[12:21]
  location= job.find('ul', class_='top-jd-dtl clearfix').text[34:]

  print(f'''
  Job_type:{job_type.strip()}
  Company Name: {comp_name.strip()}
  Skills: {skills.strip()}
  Job Published Date: {job_pub_date.strip()}
  More Info: {More_info}
  Year:{year.strip()}
  Location:{location.strip()}
  ''')
```

```python
#Writing a text file
for index,job in enumerate(jobs):
  comp_name= job.find('h3', class_ = 'joblist-comp-name').text
  skills = job.find('span', class_='srp-skills').text.replace(' ','')
  job_pub_date = job.find('span', class_= 'sim-posted').text
  More_info = job.header.h2.a['href']

  with open (f' posts.txt', 'w') as f:
    f.write (f" Company Name: {comp_name.strip()} \n")
    f.write (f" Skills: {skills.strip()}\n")
    f.write (f" Job Published Date: {job_pub_date.strip()}\n")
    f.write (f" More Info: {More_info}\n")
```

```python
#Creating a data frame for the 1st page
import pandas as pd

whole_info= []
for index,job in enumerate(jobs):
  job_type= job.header.h2.a.text
  comp_name= job.find('h3', class_ = 'joblist-comp-name').text
  skills = job.find('span', class_='srp-skills').text.replace(' ','')
  job_pub_date = job.find('span', class_= 'sim-posted').text
  More_info = job.header.h2.a['href']
  year = job.find('ul', class_='top-jd-dtl clearfix').text[12:21]
  location= job.find('ul', class_='top-jd-dtl clearfix').text[34:]

  job_info = {
      'index':index,
      'Job_Type':job_type.strip(),
      'Company_Name': comp_name.strip(),
       'Required_Skills': skills.strip(),
       'Years_of_experience':year.strip(),
       'Job_Location':location.strip(),
       'Job_published_date':job_pub_date.strip(),
       'More_info': More_info.strip()}
  whole_info.append(job_info)

df = pd.DataFrame(whole_info)
df.head(10)
```

```python
#Creating a data frame for the first 10 pages
import pandas as pd
whole_info= []

for i in range(1,10):
  url= 'https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=python&pDate=I&sequence='
  r = requests.get(url+str(i)+'&startPage=')
  soup = BeautifulSoup(r.content, 'html.parser')
  jobs = soup.find_all('li', class_='clearfix job-bx wht-shd-bx') 
  
  for job in jobs:
    #Job Type
    jtype= job.header.h2.a
    if jtype is not None: job_type = jtype.text
    else: jtype = None

    #Required Year
    jyear = job.find('ul', class_='top-jd-dtl clearfix')
    if jyear is not None: year = jyear.text[12:21]
    else: jyear = None

    #Job Location
    jlocation= job.find('ul', class_='top-jd-dtl clearfix')
    if jlocation is not None: location = jlocation.text[34:]
    else: jlocation = None

    #Company Name
    comp = job.find('h3', class_ = 'joblist-comp-name')
    if comp is not None: comp_name = comp.text
    else: comp = None
    
    #Skills
    skill = job.find('span', class_='srp-skills')
    if skill is not None: skills = skill.text
    else: skill = None

    #Job publish date
    pub_date = job.find('span', class_= 'sim-posted')
    if pub_date is not None: job_pub_date = pub_date.text
    else: pub_date = None

    job_info = {
      'index':index,
      'Job_Type':job_type.strip(),
      'Company_Name': comp_name.strip(),
       'Required_Skills': skills.strip(),
       'Years_of_experience':year.strip(),
       'Job_Location':location.strip(),
       'Job_published_date':job_pub_date.strip()}
       
    whole_info.append(job_info)

```

```python
df = pd.DataFrame(whole_info)
df.head()
```
